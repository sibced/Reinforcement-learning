```python
from ppo import do_ppo
from ppo2 import do_our_ppo
from models import PolicyContinuousModel, PolicyDiscreteModel, ValueModel
```

```python
do_ppo(
    PolicyContinuousModel(),
    ValueModel(),
    "PPO-Continuous/",
    n_actors=50,
    max_len=200,
    n_epochs=10,
    batch_size=64,
    n_iteration=50,
    eps=0.2,
)
```

```python
do_ppo(
    PolicyDiscreteModel([-1.0, -0.5, -0.1, 0.0, 0.1, 0.5, 1.0]),
    ValueModel(),
    "PPO-Discrete/",
    n_actors=50,
    max_len=200,
    n_epochs=10,
    batch_size=64,
    n_iteration=50,
    eps=0.2,
)
```
